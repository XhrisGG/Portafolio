# **Aula 5 - Leyendo bancos de datos**

## **Creando un banco de datos local**

```python
import sqlalchemy
```

```python
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine ('sqlite:///:memory:')
```

## **Escribiendo en un banco de datos**

```python
import pandas as pd
```

```python
archivo_bd = '/content/clientes_banco.csv'
```

```python
datos_bd = pd.read_csv(archivo_bd)
```

```python
datos_bd.head(5)
```

```python
datos_bd.to_sql('clientes_bd', engine, index=False)
```

```python
inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

## **Leyendo una consulta SQL**

```python
query ='SELECT * FROM clientes_bd WHERE Categoria_de_renta = "Empleado"'
```

```python
empleados = pd.read_sql(sql=text(query), con = engine.connect())
```

```python
empleados
```

```python
empleados.to_sql('empleados', con = engine.connect(),index=False)
```

```python
pd.read_sql_table('empleados', con = engine.connect())
```

```python
pd.read_sql_table('empleados', con = engine.connect(), columns = ['ID_Cliente','Grado_estudio','Rendimiento_anual'])
```

## **Actualizando un banco de datos**

```python
query = 'SELECT * FROM clientes_bd'
```

```python
pd.read_sql(sql=text(query), con = engine.connect())
```

```python
# Borrando registros
from sqlalchemy.exc import SQLAlchemyError
query = "DELETE FROM clientes_bd WHERE ID_Cliente = 5008804"
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registro borrado: ", r_set.rowcount)
```

```python
pd.read_sql_table('clientes_bd', con = engine.connect())
```

```python
# Actualizando registros
query = "UPDATE clientes_bd SET Grado_estudio='Nivel superior' WHERE ID_Cliente = 5008808"
try:
  r_set = engine.connect().execute(text(query))
except SQLAlchemyError as e:
  print(e)
else:
  print("#Registro actualizados: ", r_set.rowcount)
```

```python
pd.read_sql_table('clientes_bd', con = engine.connect())
```

```python
inspector = inspect(engine)
```
